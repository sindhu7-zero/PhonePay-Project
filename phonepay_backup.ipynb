```python
import pandas as pd
import json
import os
```

```python
agg_ins_statepath=r"C:\Users\admin\Documents\phonepay_project\pulse\data\aggregated\insurance\country\india\state"
agg_ins_statelist=os.listdir(agg_ins_statepath)
```

```python
agg_ins_data={'state':[],
              'year':[],
              'quarter':[],
              'transaction_type':[],
              'transaction_count':[],
              'transaction_amount':[]}

for state in agg_ins_statelist:
    years=os.listdir(f'{agg_ins_statepath}/{state}')
    for year in years:
        quarter=os.listdir(f'{agg_ins_statepath}/{state}/{year}')
        for qtr in quarter:
            path=(f'{agg_ins_statepath}/{state}/{year}/{qtr}')

            with open(path,'r') as f:
                data=json.load(f)
        
            for i in data['data']['transactionData']:
                transaction_type=i['name']
                count=i['paymentInstruments'][0]['count']
                amount=i['paymentInstruments'][0]['amount']
            #print(district,count,amount)
            #print(district,count,amount,state,year,qtr)
                agg_ins_data['transaction_type'].append(transaction_type)
                agg_ins_data['transaction_count'].append(count)
                agg_ins_data['transaction_amount'].append(amount)
                agg_ins_data['state'].append(state)
                agg_ins_data['year'].append(year)
                agg_ins_data['quarter'].append(int(qtr.strip('.json')))
 
agg_ins_df=pd.DataFrame(agg_ins_data)
agg_ins_df
agg_ins_df.to_sql(name='aggregate_insurance',con=engine,if_exists='append', index=False)
```
